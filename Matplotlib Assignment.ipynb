Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

#Ans_2.
* A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional (2D) plane, typically with one variable plotted on the x-axis and another variable plotted on the y-axis. Each dot on the scatter plot represents a single data point, and the position of the dot is determined by the values of the two variables it represents.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is my X data")
plt.ylabel("This is my Y data")
plt.title("This is my Graph Ploting")
```

#Ans_3.
* The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts side by side or in a grid layout, which can be helpful for comparing different data visualizations or presenting related information.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
ax1 = plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('Line 1')
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
ax2 = plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.title('Line 2')
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
ax3 = plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Line 3")
plt.show()
```

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
ax4 = plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Line 4")
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

#Ans_4.
* A bar plot, also known as a bar chart or bar graph, is a data visualization that represents categorical data using rectangular bars. Each bar's length or height is proportional to the value it represents, making it a valuable tool for visualizing and comparing the values of different categories or groups. Bar plots are widely used in data analysis and visualization

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

#Ans_5.
* Box plots provide a quick visual summary of the variability of values in a dataset. They show the median, upper and lower quartiles, minimum and maximum values, and any outliers in the dataset. Outliers can reveal mistakes or unusual occurrences in data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig,ax = plt.subplots()
ax.boxplot([box1, box2])
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot')
ax.set_xlabel('Boxes')
ax.set_ylabel('Values')
plt.show()
```
